```python
# Function to display icons in the tkinter window
def display_icons(vis_df, root):
    for index, row in vis_df.iterrows():
        title = row['Title']
        neuroimaging_data = row['neuroimaging']
        nm_list = neuroimaging_data.split(",")

        title_label = ctk.CTkLabel(root, text=title, font=("Arial", 14))
        title_label.grid(row=index, column=0, padx=10, pady=10, sticky="w")
        
        for i, nm in enumerate(nm_list):
            icon_name = icon_mapping[nm.strip()]
            icon = load_icon(icon_name)
            icon = icon.resize((50, 50))  # Resize for display
            ctk_icon = ctk.CTkImage(dark_image=icon, size=(50, 50))
            
            icon_label = ctk.CTkLabel(root, image=ctk_icon)
            icon_label.image = ctk_icon  # Keep a reference to avoid garbage collection
            icon_label.grid(row=index, column=i + 1, padx=5, pady=5)
```
