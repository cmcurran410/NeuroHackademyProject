```python
customtkinter.set_appearance_mode("dark")
customtkinter.set_default_color_theme("dark-blue")

class App(customtkinter.CTk):
    def __init__(self):
        super().__init__()

        self.title("Test TLDR App")
        self.geometry("1000x500")

        icon_dir = "'\\Users\\loret\\OneDrive\\Documents\\MRI\\Neurohackademy_2024\\yenv\\neuroimaging_icons'"
        icon = icon.open(icon_dir)

        icon = icon.resize((100, 100), Image.ANTIALIAS)
        tk_image = customtkinter.CTKImage(icon)

        self.icon_label = customtkinter.CTKLabel(self, icon=tk_image, text="") 
        self.icon_label.image = tk_image
        self.icon_label.grid(row=4, column=0, padx=10, pady=10, sticky="ew")

        for i in range(6):
            self.grid_rowconfigure(i, weight=1)
            self.grid_columnconfigure(i, weight=1)
        self.plot_graph()
        self.assignIcon(self)

        label1 = customtkinter.CTkLabel(master=self, text="Descriptive Statistics", text_color="pink", font=("Roboto", 20))
        label1.grid(row=0, column=0, padx=10, pady=10, sticky="ew", columnspan=6)

        label2 = customtkinter.CTkLabel(master=self, text="Title", font=("Roboto", 16))
        label2.grid(row=1, column=2, padx=10, pady=10, sticky="ew")

        label3 = customtkinter.CTkLabel(master=self, text="Graph", font=("Roboto", 16))
        label3.grid(row=2, column=4, padx=10, pady=10, sticky="ew", columnspan=2)

        #self.button1 = customtkinter.CTkButton(self, text="Descriptive Statistics", command=self.button_callback)
        #self.button1.grid(row=0, column=0, padx=10, pady=10, sticky="ew", columnspan=6)
        #self.button1.configure(fg_color="green")

        #self.button2 = customtkinter.CTkButton(self, text="Title", height=20, width=40)
        #self.button2.grid(row=1, column=2, padx=10, pady=10, sticky="ew")
        #self.button2.configure(fg_color="purple")
        
        #self.button3 = customtkinter.CTkButton(self, text="Stacked_Graph", height=20, width=20, command=self.button_graph)
        #self.button3.grid(row=2, column=4, padx=10, pady=10, sticky="ew", columnspan=2)
        #self.button3.configure(fg_color="blue")

    def plot_graph(self):
        # Create a Matplotlib figure and axis
        fig, ax = plt.subplots(figsize=(8, 6))
        vis_df = pd.read_excel('sample_dataFrame.xlsx')

        # Example data
        x = vis_df["sex"]
        y = vis_df["sample_size"]
        
        # Plot the data
        ax.plot(x, y, label="Sample Data")
        ax.set_xlabel("X Axis")
        ax.set_ylabel("Y Axis")
        ax.set_title("Descriptive Plot")
        ax.legend()

        canvas = FigureCanvasTkAgg(fig, master=self)  # A tk.DrawingArea
        canvas.draw()
        
        # Position the canvas on the Tkinter window
        canvas.get_tk_widget().grid(row=3, column=4, columnspan=3, padx=10, pady=10, sticky="s")
        
    def load_icon(self, icon_name):
        icon_path = os.path.join(self.icon_dir, f'{icon_name}.png')
        print(f"Loading icon from: {icon_path}")
        return Image.open(icon_path)

    def plot_nm_w_icons(self, title, neuroimaging_data):
        nm_list = neuroimaging_data.split(",")
        nm_count = len(nm_list)
        print(nm_list)
        
        fig, ax = plt.subplots(1, nm_count, figsize = (2 * nm_count, 2), tight_layout = True)
        fig.suptitle("Neuroimaging Modalities", fontsize = 12)
        for i, nm in enumerate(nm_list): # for each imaging modality
            icon = self.load_icon(self.icon_mapping[nm]) 
            if nm_count > 1:   
                ax_handle = ax.flat[i]
            else: 
                ax_handle = ax
            ax_handle.imshow(icon, aspect = "auto")
            ax_handle.axis("off")
        print()
    
    def assignIcon(self):
        vis_df = pd.read_excel('sample_dataFrame.xlsx')
        for index, row in vis_df.iterrows():
            title = row['Title']
            neuroimaging_data = row['neuroimaging']
            print(title)
            print(neuroimaging_data)
            self.plot_nm_w_icons(self, title, neuroimaging_data)
            canvas = FigureCanvasTkAgg(fig, master=self)  # A tk.DrawingArea
            canvas.draw()
        
        # Position the canvas on the Tkinter window
        canvas.get_tk_widget().grid(row=3, column=4, columnspan=3, padx=10, pady=10, sticky="s")

if __name__ == "__main__":
    app = App()
    app.mainloop()
```
